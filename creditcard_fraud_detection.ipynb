# Credit Card Fraud Detection

```python
# Importing relevant libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
credit_card_data = pd.read_csv('creditcard.csv')
```

- There are 284315 normal transactions and 492 fraudulent transactions
- Dataset is highly imbalanced.

```python
# Separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

- The difference in the mean values between groups will be reall beneficial in the machine learning part.

## Under-Sampling

- Building a sample dataset containing similar distribution of legit transactions and fraudulent transactions
- Number of fraudulent transactions --> 492

```python
# taking random 492 points from the legit variable that contains similar distribution
legit_sample = legit.sample(n = 492)
```

```python
# concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
new_dataset.groupby('Class').mean()
```

- Our new dataset contains similar distribution as shown above, '.sample' function works like a charm.

## Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
# Splitting data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
# accuracy on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
# accuracy on testing data
X_test_pred = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print('Accuracy on testing data: ', test_data_accuracy)
```

- The model works perfectly well.
